* CHECKING NULL VALUES AND HANDLING THEM :

```python
rows_with_nan = consumers[consumers.isnull().any(axis=1)]
rows_with_nan
```

```python
consumers['Smoker']= consumers['Smoker'].fillna('No')
```

```python
consumers["Transportation_Method"] = consumers["Transportation_Method"].fillna('Public')
```

```python
consumers['Marital_Status'] = consumers['Marital_Status'].fillna('Single')
```

```python
consumers['Children']=consumers['Children'].fillna('Independent')
```

```python
consumers['Occupation'] = consumers.apply(
    lambda row: 'Student' if pd.isna(row['Occupation']) and row['Age'] <= 23 else row['Occupation'],
    axis=1
)
```

```python
consumers['Occupation'] = consumers.apply(
    lambda row: 'Employed' if pd.isna(row['Occupation']) and row['Age'] >= 23 else row['Occupation'],
    axis=1
)
```

```python
consumers['Budget'] = consumers['Budget'].fillna('Medium')
```
